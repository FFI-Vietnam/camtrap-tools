```python
# @title List of next upload animal images

# List of currenly unrecored species and need to be ignored
unrecorded_species = ["Hill Blue Flycatcher", "Indochinese Wren-babbler", "Small Asian Mongoose", "Flying Insect", \
                      "Natalia's Agama", "Black-throated Laughingthrush", "Banded Krait", "Impressed Tortoise", "Animal"]

species_counts = pd.read_csv('species_counts.csv')
metadata_remove_blank = pd.read_csv('metadata_remove_blank.csv')

image_animal_id = []
animal = []

for species in species_counts['species_common_name']:
  # find # of images of a particular species to upload, i.e. 80% of total
  if species not in unrecorded_species:
    num_new_upload = round((species_counts[species_counts['species_common_name'] == species]['counts']*0.8 - \
                   species_counts[species_counts['species_common_name'] == species]['uploaded']).iat[0])
    if num_new_upload > 0:
      # select unselected images which has the species
      list_of_files = metadata_remove_blank[(metadata_remove_blank['species_common_name'] == species) & \
                                      (metadata_remove_blank['select'] != 'x')].head(num_new_upload)['FileName'].to_list()
      image_animal_id += list_of_files
      animal+=[species]*len(list_of_files)
                
# print(image_id)
print(len(image_animal_id))
print('Supposed to be', (np.sum(species_counts['counts'])*0.8-np.sum(species_counts['uploaded'])))

```

```python
#@title List of next upload blank images

metadata_select = pd.read_csv('metadata_select.csv')

next_blank = metadata_select[metadata_select['species_common_name'].isna() & metadata_select['people'].isna()]
next_blank = next_blank[next_blank['select'] != 'x']

image_blank_id = []
blank = []

image_blank_id += next_blank.sample(int(0.8*len(next_blank)))['FileName'].to_list()
blank += ['blank']*int(0.8*len(next_blank))


print(len(blank))
```

```python
#@title Save to next_upload.csv

image_id = []
image_id += image_animal_id
image_id += image_blank_id
tag = []
tag += animal
tag += blank

next_upload = pd.DataFrame()
next_upload['filename'] = image_id
next_upload['tag'] = tag

print(len(next_upload))

next_upload.to_csv('next_upload.csv', index = False)
print('Saved file to next_upload.csv')
next_upload.sample(10)
```

```python
#@title Code image template and save to images_filled.csv

images_template = pd.read_csv('images.csv')
taxon_match_table = pd.read_csv('taxon_match_table.csv')
next_upload = pd.read_csv('next_upload.csv')

project_id = []
deployment_id = []
location = []
identified_by = []
wi_taxon_id = []
class_taxon = []
order_taxon = []
family_taxon = []
genus_taxon = []
species_taxon = []
common_name = []
timestamp = []
number_of_objects = []

#--------NOT REQUIRED--------
uncertainty = []
highlighted = []
age = []
sex = []
animal_recognizable = []
individual_id = []
individual_animal_notes = []
markings = []
external_sequence_id = []

#--------AUTOFILL EACH FIELDS--------
for j in range(len(image_id)):
  i = image_id[j]
  project_id.append(projects_template['project_id'][0])
  deployment_id.append(f'KonPlong1920-{i.split("__")[0]}-{i.split("__")[1]}')
  location.append(metadata_remove_blank[metadata_remove_blank['FileName'] ==  i]['SourceFile'].iat[0])
  identified_by.append(metadata_remove_blank[metadata_remove_blank['FileName'] ==  i]['observer'].iat[0])
  # print(metadata_remove_blank[metadata_remove_blank['FileName'] ==  i]['species_common_name'].iat[0])

  sub_wi_taxon_id = taxon_match_table[
                                      taxon_match_table['FFI_species_name'] == 
                                      metadata_remove_blank[(metadata_remove_blank['FileName'] ==  i) & 
                                                            (metadata_remove_blank['species_common_name'] ==  next_upload[next_upload['filename']==i]['tag'].iat[0])]['species_common_name'].iat[0]
                                      ]
    
  wi_taxon_id.append(sub_wi_taxon_id["uniqueIdentifier"].iat[0])
  class_taxon.append(sub_wi_taxon_id["class"].iat[0])
  order_taxon.append(sub_wi_taxon_id["order"].iat[0])
  family_taxon.append(sub_wi_taxon_id["family"].iat[0])
  genus_taxon.append(sub_wi_taxon_id["genus"].iat[0])
  species_taxon.append(sub_wi_taxon_id["species"].iat[0])
  common_name.append(sub_wi_taxon_id["commonNameEnglish"].iat[0])

  timestamp.append(extract_timestamp_from_FileName(i))

  number_of_objects.append("")
  uncertainty.append("")
  highlighted.append("")
  age.append("")
  sex.append("")
  animal_recognizable.append("")
  individual_id.append("")
  individual_animal_notes.append("")
  markings.append("")
  external_sequence_id.append("")


#--------AUTOFILL TEMPLATES--------
images_template["project_id"] = project_id
images_template["deployment_id"] = deployment_id
images_template["image_id"] = image_id
images_template["location"] = location
images_template["identified_by"] = identified_by
images_template["wi_taxon_id"] = wi_taxon_id
images_template["class"] = class_taxon
images_template["order"] = order_taxon
images_template["family"] = family_taxon
images_template["genus"] = genus_taxon
images_template["species"] = species_taxon
images_template["common_name"] = common_name
images_template["uncertainty"] = uncertainty
images_template["timestamp"] = timestamp
images_template["number_of_objects"] = number_of_objects
images_template["highlighted"] = highlighted
images_template["age"] = age
images_template["sex"] = sex
images_template["animal_recognizable"] = animal_recognizable
images_template["individual_id"] = individual_id
images_template["individual_animal_notes"] = individual_animal_notes
images_template["markings"] = markings
images_template["external_sequence_id"] = external_sequence_id

# Save to file
images_template.to_csv('images_filled.csv', index = False)
print('Saved to images_filled.csv')
images_template
```
