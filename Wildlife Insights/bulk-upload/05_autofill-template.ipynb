```python
# useful functions
def read_csv_Google_drive(root, contain_folder, file_name):
  """
  function to read a csv file from Google Drive
  param examples:
    root = '/content/drive/'
    contain_folder = 'My Drive/FFI/dataset'
    file_name = 'image_metadata(2020-06-26)_full.csv'
  """
  file_path = os.path.join(root, contain_folder, file_name)
  return pd.read_csv(file_path)

def save_csv_Google_drive(df, root, contain_folder, file_name):
  """
  function to save a csv file to Google Drive
  param examples:
    root = '/content/drive/'
    contain_folder = 'My Drive/FFI/dataset'
    file_name = 'image_metadata(2020-06-26)_full.csv'
  """
  # save file to Colab runtime storage (will be deleted when this notebook is closed)
  df.to_csv('dataframe.csv', index=False)

  # save file back to Google Drive for permanent storage
  folder_path = os.path.join(root, contain_folder)
  file_path = os.path.join(root, contain_folder, file_name)
  try:
    os.makedirs(folder_path)
  except:
    pass

  with open('dataframe.csv', 'r') as f:
    df_file = f.read()

  with open(file_path, 'w') as f:
    f.write(df_file)

  print(f'File is saved to {file_name} in Google Drive at {file_path}')
```

```python
# deployments template

# read template file
deployments_template = read_csv_Google_drive(root, templates_folder, "deployments.csv")

# read clean dataset
konplong = read_csv_Google_drive(root, contain_folder, "2.3_clean_deployment_format-datetime.csv")
konplong['Camera'] = konplong['Camera'].astype('int')

project_id = [""]
deployment_id = [""]
placename = [""]
longitude = [""]
latitude = [""]
start_date = [""]
end_date = [""]
bait_type = [""]
feature_type = [""]
camera_id = [""]
quiet_period = [""]
camera_functioning = [""]
sensor_height = [""]
sensor_orientation = [""]
# not required
subproject_name = [""]
subproject_design = [""]
event_name = [""]
event_description = [""]
event_type = [""]
bait_description = [""]
feature_type_methodology = [""]
height_other = [""]
orientation_other = [""]
recorded_by = [""]
plot_treatment = [""]
plot_treatment_description = [""]
detection_distance = [""]

# fill in info
for i in range(len(konplong)):
  project_id.append(projects_template['project_id'][0])
  deployment_id.append('KONPLONG1920-' + str(konplong['Station'][i]).zfill(3) + "-" + str(konplong['Camera'][i]).zfill(5))
  placename.append('KONPLONG1920-' + str(konplong['Station'][i]).zfill(3) + "-" + str(konplong['Camera'][i]).zfill(5))
  longitude.append(konplong['Longitude'][i])
  latitude.append(konplong['Latitude'][i])
  start_date.append(konplong['Datetime setup'][i])
  if konplong['Datetime retrieval'][i] == 'error':
    end_date.append(konplong['Datetime for null camera-off'][i])
    camera_functioning.append("Memory Card Failure")
  else:
    end_date.append(konplong['Datetime retrieval'][i])
    if konplong['Camera still functioning at retrieval (Y/N)?'][i] == 'N':
      camera_functioning.append("Unknown Failure")
    else:
      camera_functioning.append("Camera Functioning")
  
  bait_type.append("None")
  feature_type.append("None")
  camera_id.append("CAM" + str(konplong['Camera'][i]))
  quiet_period.append("0")
  sensor_height.append("Knee height")
  sensor_orientation.append("Parallel")
  subproject_name.append(f"Phase {str(konplong['Deployment Phase'][i]).zfill(2)}")

  subproject_design.append("")
  event_name.append("")
  event_description.append("")
  event_type.append("")
  bait_description.append("")
  feature_type_methodology.append("")
  height_other.append("")
  orientation_other.append("")
  recorded_by.append("")
  plot_treatment.append("")
  plot_treatment_description.append("")
  detection_distance.append("")

# create dataframe
deployments_template["project_id"] = project_id
deployments_template["deployment_id"] = deployment_id
deployments_template["subproject_name"] = subproject_name
deployments_template["subproject_design"] = subproject_design
deployments_template["placename"] = placename
deployments_template["longitude"] = longitude
deployments_template["latitude"] = latitude
deployments_template["start_date"] = start_date
deployments_template["end_date"] = end_date
deployments_template["event_name"] = event_name
deployments_template["event_description"] = event_description
deployments_template["event_type"] = event_type
deployments_template["bait_type"] = bait_type
deployments_template["bait_description"] = bait_description
deployments_template["feature_type"] = feature_type
deployments_template["feature_type_methodology"] = feature_type_methodology
deployments_template["camera_id"] = camera_id
deployments_template["quiet_period"] = quiet_period
deployments_template["camera_functioning"] = camera_functioning
deployments_template["sensor_height"] = sensor_height
deployments_template["height_other"] = height_other
deployments_template["sensor_orientation"] = sensor_orientation
deployments_template["orientation_other"] = orientation_other
deployments_template["recorded_by"] = recorded_by
deployments_template["plot_treatment"] = plot_treatment
deployments_template["plot_treatment_description"] = plot_treatment_description
deployments_template["detection_distance"] = detection_distance

# save to Google Drive
save_csv_Google_drive(deployments_template, root, contain_folder, deployments_file_name)

deployments_template.sample(5)
```

```python
# images template 

def to_GCP_location(source_file, parents_folder, GCP_bucket_folder):
  """ 
  change file location to point to GCP bucket
  ex: 
    source_file: D:/Fauna & Flora International/VietnamCameraStorage - Root/Kon Plong/2019-2020 Survey/Processed images/Final/100/67096/100__67096__2019-04-22__15-09-11(1).JPG
    parents_folder: D:/Fauna & Flora International/VietnamCameraStorage - Root/Kon Plong/2019-2020 Survey/Processed images/Final
    GCP_bucket_folder: gs://ffi_vietnam/KONPLONG19-20
    --> return: gs://ffi_vietnam/KONPLONG19-20/100/67096/100__67096__2019-04-22__15-09-11(1).JPG
  """
  return source_file.replace(parents_folder, GCP_bucket_folder)

images_template = read_csv_Google_drive(root, templates_folder, 'images.csv')
taxon_match_table = read_csv_Google_drive(root, contain_folder, "4.1_collect_WI-taxon-database.csv")
metadata_single_species = read_csv_Google_drive(root, contain_folder, "1.3_clean_metadata_single-species-row.csv")

image_id = []
project_id = []
deployment_id = []
location = []
identified_by = []
wi_taxon_id = []
class_taxon = []
order_taxon = []
family_taxon = []
genus_taxon = []
species_taxon = []
common_name = []
timestamp = []
number_of_objects = []
# not required
uncertainty = []
highlighted = []
age = []
sex = []
animal_recognizable = []
individual_id = []
individual_animal_notes = []
markings = []
external_sequence_id = []

parents_folder = "D:/Fauna & Flora International/VietnamCameraStorage - Root/Kon Plong/2019-2020 Survey/Processed images/Final"
GCP_bucket_folder = "gs://ffi_vietnam/KONPLONG19-20"

unspecified_species = ["nan", 
                       "Black-throated Laughingthrush", 
                       "Impressed Tortoise",
                       "Banded Krait", 
                       "Small Asian Mongoose", 
                       "Natalia's Agama"
                       ]

# fill in info
for i in range(len(metadata_single_species)):
  image_id.append(metadata_single_species.loc[i, 'FileName'])
  project_id.append(projects_template['project_id'][0])
  deployment_id.append('KONPLONG1920-' + str(metadata_single_species.loc[i, 'station']).zfill(3) + "-" + str(metadata_single_species.loc[i, 'camera_id']).zfill(5))
  location.append(to_GCP_location(metadata_single_species.loc[i, 'SourceFile'], parents_folder, GCP_bucket_folder))
  identified_by.append(metadata_single_species.loc[i, 'observer'])

  if str(metadata_single_species.loc[i, 'species_common_name']) in unspecified_species: # if blank image
    sub_wi_taxon_id = taxon_match_table[taxon_match_table['FFI_species_name'] == 'Blank']
    number_of_objects.append("")
  else:
    sub_wi_taxon_id = taxon_match_table[taxon_match_table['FFI_species_name'] == metadata_single_species.loc[i, 'species_common_name']]
    number_of_objects.append("1")
  
  wi_taxon_id.append(sub_wi_taxon_id["uniqueIdentifier"].iat[0])
  class_taxon.append(sub_wi_taxon_id["class"].iat[0])
  order_taxon.append(sub_wi_taxon_id["order"].iat[0])
  family_taxon.append(sub_wi_taxon_id["family"].iat[0])
  genus_taxon.append(sub_wi_taxon_id["genus"].iat[0])
  species_taxon.append(sub_wi_taxon_id["species"].iat[0])
  common_name.append(sub_wi_taxon_id["commonNameEnglish"].iat[0])

  timestamp.append(metadata_single_species.loc[i, 'datetime_created'])

  uncertainty.append("")
  highlighted.append("")
  age.append("")
  sex.append("")
  animal_recognizable.append("")
  individual_id.append("")
  individual_animal_notes.append("")
  markings.append("")
  external_sequence_id.append("")


# create dataframe
images_template["project_id"] = project_id
images_template["deployment_id"] = deployment_id
images_template["image_id"] = image_id
images_template["location"] = location
images_template["identified_by"] = identified_by
images_template["wi_taxon_id"] = wi_taxon_id
images_template["class"] = class_taxon
images_template["order"] = order_taxon
images_template["family"] = family_taxon
images_template["genus"] = genus_taxon
images_template["species"] = species_taxon
images_template["common_name"] = common_name
images_template["uncertainty"] = uncertainty
images_template["timestamp"] = timestamp
images_template["number_of_objects"] = number_of_objects
images_template["highlighted"] = highlighted
images_template["age"] = age
images_template["sex"] = sex
images_template["animal_recognizable"] = animal_recognizable
images_template["individual_id"] = individual_id
images_template["individual_animal_notes"] = individual_animal_notes
images_template["markings"] = markings
images_template["external_sequence_id"] = external_sequence_id

# save list of unspecified species by WI 
save_csv_Google_drive(metadata_single_species[metadata_single_species['species_common_name'].isin(unspecified_species[1:])], root, contain_folder, unspecified_species_file_name)

# save to Google Drive
save_csv_Google_drive(images_template, root, contain_folder, images_file_name)
images_template.sample(10)
```

```python
# sanity check images.csv
# the sample is from previous web upload with around 3000 images. I use this sample
# to random check if the images.csv is filled with correctness

sample_folder  = 'My Drive/FFI/Wildlife Insights Bulk Upload Test/bulk-upload_template-autofill/Kon Plong sample templates'
sample_images_template = read_csv_Google_drive(root, sample_folder, "images.csv")
sample_images_template = sample_images_template[['filename', 'wi_taxon_id', 'class', 'genus', 'species', 'common_name', 'timestamp']]

test_images_template = images_template[['image_id', 'wi_taxon_id', 'class', 'genus', 'species', 'common_name', 'timestamp']]
# only get the files that appear in the sample dataframe
test_images_template = test_images_template[test_images_template['image_id'].isin(sample_images_template['filename'].to_list())]

count = 0 # number of differences
for i in range(len(test_images_template['image_id'])):
  filename = test_images_template['image_id'].to_list()[i]
  test_row = test_images_template.iloc[i].to_list()
  sample_row = sample_images_template[sample_images_template['filename'] == filename].iloc[0].to_list()
  if not test_row == sample_row:
    count += 1
    print("test  ", test_row)
    print("sample", sample_row)
    print()
print('There are', count, 'differences in', len(sample_images_template), "entries in total")
```
