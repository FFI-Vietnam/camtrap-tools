```python
def create_dat_labs(threshold):
  dat = read_csv_Google_drive(root, contain_folder, confusion_matrix_file_name)
  taxon_match_table = read_csv_Google_drive(root, contain_folder, taxon_match_table_file_name)
  ground_truth = read_csv_Google_drive(root, contain_folder, ground_truth_file_name)

  # unspecified species
  unspecified_species = ['Banded Krait', 
                        'Black-throated Laughingthrush',
                        'Impressed Tortoise',
                        'Small Asian Mongoose',
                        'All',
                        'Animal',
                        'Mammal',
                        'Flying Insect',
                        'Blank']

  # match order name
  order_match = {
      'Primates': 'Primates', 
      'Carnivora': 'Carnivores', 
      'Passeriformes': 'Passerine', 
      'Columbiformes': 'Doves and Pigeons',
      'Piciformes': 'Woodpeckers', 
      'Strigiformes': 'Owls', 
      'Rodentia': 'Rodents', 
      'Cuculiformes': 'Cuckoos',
      'Scandentia': 'Treeshrew', 
      'Lepidoptera': 'Butterflies and Moths', 
      'Chiroptera': 'Bats', 
      'Galliformes': 'Landfowls',
      'Coraciiformes': 'Coraciiformes', 
      'Artiodactyla': 'Even-toed Ungulates', 
      'Cetartiodactyla': 'Even-toed Ungulates',
      'Accipitriformes': 'Eagles', 
      'Squamata': 'Lizards and Snakes', 
      'Gruiformes': 'Crane-like Birds', 
      'Eulipotyphla': 'Eulipotyphla',
      'Charadriiformes': 'Shorebirds', 
      'Pholidota': 'Pangolins', 
      'Pelecaniformes': 'Pelicans',
      'Hello': 'Hello'
  }

  # match class name
  class_match = {
      'Mammalia': 'Mammals', 
      'Aves': 'Birds', 
      'Reptilia': 'Reptiles'
  }

  # change and drop columns
  dat['FFI_species_name'] = dat['Unnamed: 0']
  dat = dat[['FFI_species_name', 'Recall']]

  # drop rows
  dat = dat[~ dat['FFI_species_name'].isin(unspecified_species)]

  # get labels
  labs = dat['Recall']
  labs = labs > THRESHOLD

  # fill na, for no order such as Birds, i.e. class Aves
  dat.fillna(value={'order':'Hello'}, inplace=True)
  taxon_match_table.fillna(value={'order':'Hello'}, inplace=True)

  # add conservation status
  # conservation_status = []
  # for species in dat['FFI_species_name']:
  #   conservation_status.append(taxon_match_table[taxon_match_table['FFI_species_name'] == species]['conservation_status'].iat[0])
  # dat['conservation_status'] = conservation_status

  # add class
  taxon_class = []
  for species in dat['FFI_species_name']:
    taxon_class.append(class_match[taxon_match_table[taxon_match_table['FFI_species_name'] == species]['class'].iat[0]])
  dat['class'] = taxon_class

  # add order
  taxon_order = []
  for species in dat['FFI_species_name']:
    taxon_order.append(order_match[taxon_match_table[taxon_match_table['FFI_species_name'] == species]['order'].iat[0]])
  dat['order'] = taxon_order

  # add day / night
  day = []
  night = []
  for species in dat['FFI_species_name']:
    flash = ground_truth[ground_truth['species_common_name'] == species]['Flash']
    if (flash == 'Fired').sum() > len(flash)/2:
      night.append(1)
      day.append(0)
    else:
      night.append(0)
      day.append(1)
  dat['diurnal'] = day
  dat['nocturnal'] = night


  # drop columns
  dat.drop(columns=['FFI_species_name', 'Recall'], inplace=True)

  # get dummies
  dat = pd.get_dummies(dat, columns=[
                                    #  'conservation_status', 
                                     'class', 
                                     'order'
                                     ])

  return dat, labs
```

```python
# create model
def make_model(dat, labs, max_leaf_nodes):
  clf = tree.DecisionTreeClassifier(max_leaf_nodes = max_leaf_nodes)	
  clf = clf.fit(dat, labs)
  return clf
```

```python
# calculate accuracy
def calculate_accuracy(clf, dat, labs):
  correct = 0						
  for i in range(len(dat)):	
      if clf.predict([dat.iloc[i]]) == labs.iloc[i]: 
        correct = correct + 1
  print('Accuracy:', 100.0* correct / len(dat))
```

```python
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual

@interact(expected_recall=(0,100,10))

def visualize_decision_tree(expected_recall):
  expected_recall = expected_recall/100
  dat, labs = create_dat_labs(expected_recall)
  clf = make_model(dat, labs, 5)
  calculate_accuracy(clf, dat, labs)
  feature_names = []

  dot_data = tree.export_graphviz(clf, 
                                  feature_names=dat.columns,
                                  class_names=[f'Recall below {expected_recall*100}', f'Recall above {expected_recall*100}'],
                                  filled=True, 
                                  rounded=True) 
  graph = graphviz.Source(dot_data)	
  return graph	
```
